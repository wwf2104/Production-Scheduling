# Imports

```python
import pandas as pd
import datetime as dt
from shapely.geometry import Point, shape
import numpy as np
```

```python
prec = pd.read_json('Police Precincts.geojson')
```

```python
prec.head()
```

```python
prec['precinct'] = prec['features'].apply(lambda x: x['properties']['precinct'])
```

```python
prec['area_coord'] = prec['features'].apply(lambda x: shape(x['geometry']))
```

```python
prec.drop(['type','features'],axis=1,inplace=True)
```

```python
prec.head()
```

```python
df = pd.read_json('data files/all_data.json')
df
```

```python
coords = df[['lat','long']]
coords = pd.DataFrame([Point(x) for x in list(zip(coords['lat'],coords['long']))])
```

```python
coords.head()
```

```python
def find_prec(pt):
    for i in range(len(prec)):
        area = prec['area_coord'].iloc[i]
        if area.contains(pt):
            return prec['precinct'].iloc[i]
    return np.nan
```

```python
manhattan = [1,5,6,7,9,10,13,17,19,20,23,24,25,26,28,30,32,33,34]
```

```python
prec = prec[prec['precinct'].apply(lambda x: int(x) in manhattan)]
```

```python
df['precinct'] = coords[0].apply(lambda x: find_prec(x))
```

```python
df.drop(['long','lat'], axis=1, inplace=True)
df.head()
```

```python
saveme = df.copy()
```

```python
df['precinct'] = saveme['precinct']
df = df.dropna()
df.to_csv('lp_df.csv', index=False)
```
